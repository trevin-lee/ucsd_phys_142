```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
timesteps = 1000000
time = 10

time_vect = np.linspace(0, time, timesteps)
position = np.cos(time_vect)
velocity = np.diff(position) / (time / timesteps)
acceleration  = np.diff(velocity) / (time / timesteps)
```

## Plotting

```python
plt.plot(time_vect, position, label='position')
plt.plot(time_vect[1:], velocity, label='velocity')
plt.plot(time_vect[2:], acceleration, label='acceleration')

plt.legend()
plt.show()
```
